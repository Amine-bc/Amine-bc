```python
import math
import numpy as np
from matplotlib import pyplot as plt
```

```python
centre = (5,5)
PI = math.pi
n = 90
r = 8

circum_vectors = [np.array((centre[0] + math.cos(2*PI/n*x)*r, centre[1] + math.sin(2*PI/n*x)*r)) for x in range(0,n)]
```

```python
M = np.random.randint(1,5,(2,2))
print('M=',M)
transformed_eclipse =[]
for v in circum_vectors:
  transformed_eclipse.append(M @ v + (30,0))
```

```python
plt.annotate("Circle", [-3,20],color='blue')
plt.scatter([p[0] for p in circum_vectors], [p[1] for p in circum_vectors])

plt.annotate("Transformed eclipse", [65,55],color='orange')
plt.scatter([p[0] for p in transformed_eclipse], [p[1] for p in transformed_eclipse])

plt.annotate("Circle is transformed to an ellipse by linear map M", [-30, 93])
plt.xlim(-50, 150)
plt.ylim(-50, 100)
plt.show()
```
